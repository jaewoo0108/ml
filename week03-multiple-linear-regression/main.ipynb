# Week 3 : 다중 선형회귀와 벡터화

##  핵심 개념
- **다중 선형 회귀(Multiple linear regression)**: 모델이 얼마나 틀리는지 측정하는 함수
- **Vectoriaztion**: 반복문(for) 없이 데이터를 행렬 통째로 묶어 계산함으로써, 컴퓨터의 병렬 연산 기능을 활용해 빠른 처리


```python
import numpy as np

# 가중치 벡터
w = np.array([1.0, 2.0, 3.0, 4.0])
# 입력 특징 벡터
x = np.array([10, 20, 30, 40])
b = 5

print(f"w shape : {w.shape}")
print(f"x shape : {x.shape}")
```

## Loop vs 벡터화 (Vectorization)

다중 선형회귀 예측:
**f = w₁x₁ + w₂x₂ + ... + wₙxₙ + b**

- 결국 벡터 w와 x의 내적(dot product) + b와 같다
- **Loop 방식**: 곱셈을 하나씩 차례대로 더함 → n 커지면 느림
- **벡터화**: `np.dot(w, x)` 한 줄, 컴퓨터가 병렬로 한 번에 계산 → 빠름


   **-> 어느방식을 사용하든 결과는 똑같다**


```python
def predict_loop(w, x, b):
  n = w.shape[0]
  f = 0
  for j in range(n):
    f = f + w[j] * x[j]
  f = f + b
  return f

# 결과
result_loop = predict_loop(w, x, b)
print(f" loop 예측값 : {result_loop} ")
```

```python
def predict_vector(w, x, b):
  # np.dot(w,x) = w와 x의 내적
  f= np.dot(w, x) + b
  return f

result_vector = predict_vector(w, x, b)
print(f" vectorization 예측값 : {result_vector}")
```

```python
# 두 방식의 시간 측정

import time

big_w = np.random.rand(1_000_000)
big_x = np.random.rand(1_000_000)

# loop 방식 시간 측정
start = time.time()
predict_loop(big_w, big_x, b)
print(f" loop 시간 : {round(time.time() - start, 4)}초 ")

## Vectorization 방식 시간 측정
start = time.time()
predict_vector(big_w, big_x, b)
print(f" vectorization 시간 : {round(time.time() - start, 4)}초")
```

```python
# compute_gradient() 함수 구현 - 각 가중치들과 편향의 기울기 구하는 함수

def compute_gradient(X, y, w, b):
  m = X.shape[0]
  n = X.shape[1]
  dj_dw = np.zeros(n)
  dj_db = 0

  for i in range(m):
    error = np.dot(X[i],w) + b - y[i]
    for j in range(n):
      dj_dw[j] = dj_dw[j] + error * X[i,j] # 오차가 발생했을 때 각 가중치가 이번 데이터의 오차에서 기여한 정도 계산위해 특징값 곱한다
    dj_db = dj_db + error

  dj_dw = dj_dw / m
  dj_db = dj_db / m

  return dj_dw, dj_db

X = np.array([[1, 2],[3, 4]])
y = np.array([5, 6])
w = np.array([0.0, 0.0])
b = 0

dj_dw, dj_db = compute_gradient(X, y, w, b)
print(f"dj_dw: {dj_dw}")
print(f"dj_db: {dj_db}")

```
